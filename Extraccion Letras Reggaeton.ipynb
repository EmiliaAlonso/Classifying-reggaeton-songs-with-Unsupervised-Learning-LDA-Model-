```python
divs = soup.find_all("li")[3:1003]

divs[0].find("a").get("title")
```

```python
#OBTENGO LOS TITULOS DE LAS CANCIONES

titulos = list()
for i in range(len(divs)):
    titulos.append(divs[i].find("a").get("title"))
titulos
```

```python
divs[0].find("span").get_text(strip=True)
```

```python
# OBTENGO LOS ARTISTAS

artistas = list()
for i in range(len(divs)):
    artistas.append(divs[i].find("span").get_text(strip =True))

```

```python
navegador = webdriver.Chrome(executable_path=r"chromedriver")
navegador.implicitly_wait(10)

# ACEPTO COOKIES
navegador.get("https://www.letras.com/mais-acessadas/reggaeton/") 
wait = WebDriverWait(navegador, 15)
wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.css-47sehv'))).click()

letras = list()
for letra in range(len(tit)):
    sleep(1)
    navegador.find_element_by_partial_link_text(tit[letra]).click()
    letras.append(navegador.find_element(By.CSS_SELECTOR, "div.cnt-letra.p402_premium").text.split("\n"))
    navegador.get("https://www.letras.com/mais-acessadas/reggaeton/")
```

```python
#ARMO EL DATASET

import pandas as pd

df = pd.DataFrame({"Títulos": titulos[:740],"Artistas":artistas[:740], "Letras": letras})
df.head()
```
